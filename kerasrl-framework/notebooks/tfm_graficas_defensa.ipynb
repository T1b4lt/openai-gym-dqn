```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Mountaincar

```python
mountaincar_rewards_noattack = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_noattack.csv").sample(200, ignore_index=True)
```

```python
mountaincar_rewards_attack_20_pm005 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_attack_20%_+-005.csv")
mountaincar_rewards_defense_20_pm005_01_10 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_defense_20%_+-005_01_10.csv")
mountaincar_rewards_attack_20_pm01 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_attack_20%_+-01.csv")
mountaincar_rewards_defense_20_pm01_01_10 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_defense_20%_+-01_01_10.csv")

mountaincar_rewards_attack_50_pm005 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_attack_50%_+-005.csv")
mountaincar_rewards_defense_50_pm005_01_10 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_defense_50%_+-005_01_10.csv")
mountaincar_rewards_attack_50_pm01 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_attack_50%_+-01.csv")
mountaincar_rewards_defense_50_pm01_01_10 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_defense_50%_+-01_01_10.csv")

mountaincar_rewards_attack_70_pm005 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_attack_70%_+-005.csv")
mountaincar_rewards_defense_70_pm005_01_10 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_defense_70%_+-005_01_10.csv")
mountaincar_rewards_attack_70_pm01 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_attack_70%_+-01.csv")
mountaincar_rewards_defense_70_pm01_01_10 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_defense_70%_+-01_01_10.csv")

mountaincar_rewards_defense_50_pm005_01_2000 = pd.read_csv("../ejecuciones_mountaincar_tfm/mountaincar_reward_defense_50%_+-005_01_2000.csv")
```

### Analisis Non_Freeze con Mountaincar

```python
mountaincar_rewards_defense_nofreeze = pd.DataFrame()

mountaincar_rewards_defense_nofreeze["Sin ataque"] = mountaincar_rewards_noattack.episode_reward.rolling(50).mean()
mountaincar_rewards_defense_nofreeze["Ataque =50% =0.05"] = mountaincar_rewards_attack_50_pm005.episode_reward.rolling(50).mean()
mountaincar_rewards_defense_nofreeze["Ataque =50% =0.05 =0.1 =10"] = mountaincar_rewards_defense_50_pm005_01_10.episode_reward.rolling(50).mean()
mountaincar_rewards_defense_nofreeze["Ataque =50% =0.05 =0.1 =2000"] = mountaincar_rewards_defense_50_pm005_01_2000.episode_reward.rolling(50).mean()


mountaincar_rewards_defense_nofreeze.plot(y=["Sin ataque",
                               "Ataque =50% =0.05",
                               "Ataque =50% =0.05 =0.1 =10",
                               "Ataque =50% =0.05 =0.1 =2000"],
                               kind="line",
                               figsize=(20,10),
                               title="Anlisis impacto no_freeze  en recompensa con  = 50% y =0.05 Mountaincar",
                               xlabel="Episodios",
                               ylabel="Recompensa")
plt.rcParams['font.size'] = '23'
plt.legend(prop={'size': 15})
```
