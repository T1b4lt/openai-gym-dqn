```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as pyo

import random
import pickle
```

## Analisis Mountaincar

```python
rewards_noattack = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_noattack.csv").sample(200, ignore_index=True)

rewards_attack = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_attack_50%_+-01.csv")
rewards_defense = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_defense_50%_+-01_005_10.csv") # defensa kmeans 0.05 non_freeze=10 ataque 50%+-0.1

# rewards_attack = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_attack_80%_+005.csv")
# rewards_defense = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_defense_80%_+005_005_10.csv") # defensa kmeans 0.05 non_freeze=10 ataque 80%+0.05
```

```python
rewards = pd.DataFrame()
rewards["average_50_noattack"] = rewards_noattack.episode_reward.rolling(50).mean()

rewards["average_50_attack_50%_+-01"] = rewards_attack.episode_reward.rolling(50).mean()
rewards["average_50_defense_50%_+-01_005_10"] = rewards_defense.episode_reward.rolling(50).mean()

# rewards["average_50_attack_80%_+005"] = rewards_attack.episode_reward.rolling(50).mean()
# rewards["average_50_defense_80%_+005_005_10"] = rewards_defense.episode_reward.rolling(50).mean()
```

```python
rewards.plot(y=["average_50_noattack",
                "average_50_attack_50%_+-01",
                "average_50_defense_50%_+-01_005_10"],
                kind="line",
                figsize=(20,10),
                title="Reward Mountaincar",
                xlabel="Episodes",
                ylabel="Reward")
```

## Analisis Cartpole

```python
rewards_noattack = pd.read_csv("../ejecuciones_cartpole/cartpole_reward_noattack.csv").sample(200, ignore_index=True)
rewards_attack = pd.read_csv("../ejecuciones_cartpole/cartpole_reward_attack_50%_+-01.csv")
rewards_defense = pd.read_csv("../ejecuciones_cartpole/cartpole_reward_defense_50%_+-01_015_3.csv") # defensa kmeans 0.15 non_freeze=3 ataque 50%+-0.1
```

```python
rewards = pd.DataFrame()
rewards["average_50_noattack"] = rewards_noattack.episode_reward.rolling(50).mean()
rewards["average_50_attack_50%_+-01"] = rewards_attack.episode_reward.rolling(50).mean()
rewards["average_50_defense_50%_+-01_015_3"] = rewards_defense.episode_reward.rolling(50).mean()
```

```python
rewards.plot(y=["average_50_noattack",
                "average_50_attack_50%_+-01",
                "average_50_defense_50%_+-01_015_3"],
                kind="line",
                figsize=(20,10),
                title="Reward Cartpole",
                xlabel="Episodes",
                ylabel="Reward")
```
