```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as pyo

import random
import pickle
```

## Analisis Mountaincar

```python
rewards_noattack = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_noattack.csv")
rewards_attack = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_attack_50%_+-01.csv")
rewards_defense_7_especial = pd.read_csv("../ejecuciones_mountaincar/mountaincar_reward_defense_7_especial.csv") # defensa kmeans 0.05 non_freeze=10 ataque 50%+-0.1
```

```python
rewards = pd.DataFrame()
rewards["average_50_noattack"] = rewards_noattack.episode_reward.rolling(100).mean()
rewards["average_50_attack_50%_+-01"] = rewards_attack.episode_reward.rolling(100).mean()
rewards["average_50_defense_7_especial"] = rewards_defense_7_especial.episode_reward.rolling(100).mean()
```

```python
rewards.plot(y=["average_50_noattack",
                "average_50_attack_50%_+-01",
                "average_50_defense_7_especial"],
                kind="line",
                figsize=(20,10),
                title="Reward Mountaincar",
                xlabel="Episodes",
                ylabel="Reward")
```
